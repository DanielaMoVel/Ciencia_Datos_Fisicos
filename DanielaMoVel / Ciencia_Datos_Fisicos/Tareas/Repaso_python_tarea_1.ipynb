````python
import math #Importamos funciones que ocuparemos a lo largo de la tarea
````

Crearemos una función en la que el usuario tenga que escribir los valores de los coeficientes de una función cuadrática $(a,b,c)$.

 $f(x)=ax^2+bx+cx$ 
 
 usando la chicharronera 
 $x=\frac{-b\pm\sqrt{b2-4ac}}{2a}$.
 
 Sin embargo para codificar tenemos que tener en cuenta cuando la solución de la ecuación cuadrática es imaginaria. Es decir cuando $b^2-ac<0$, para este caso la ecuación mandará un mensaje de que no es posible resolver con la chicharronera. 

 Como vemos tenemos dos resultados distintos, es decir un $x_1$ en donde se usa el signo positivo de la raíz y otra solución $x_2$ para cuando usamos el signo negativo. Entonces nuestra función también debe de incluír estos casos. 

````python
def chicharronera(a,b,c):
    if ((b**2)-4*a*c) < 0:
      print("No se puede resolver para números complejos con la chicharronera")
    else:
      x1 = (-b+sqrt(b**2-(4*a*c)))/(2*a)  # Usando el signo positivo
      x2 = (-b-sqrt(b**2-(4*a*c)))/(2*a)  # Usando el signo negativo
      print("Los valores de x son:")
      print(x1)
      print(x2)
````

````python
fx = chicharronera(10,-2,0)
fx
````

Para resolver este problema, tenemos que tener en cuenta lo siguiente:

### Ejercicio 6 (2 puntos): Aproximación de $\pi$
Una de las formas más sencillas de aproximar $\pi$ es mediante la siguiente serie infinita de Leibniz:
$$\frac{\pi}{4} = \sum_{n=0}^{\infty}\frac{(-1)^{n}}{(2n+1)} = \frac{1}{1} - \frac{1}{3} + \frac{1}{5}-\frac{1}{7}+\frac{1}{9}-...$$

crea una función para aproximar $\pi$ utilizando la serie de Leibniz. Usa n = 100000 e imprime el valor de tu aproximación tanto el valor de $\frac{\pi}{4}$ como $\pi$.

El valor de $\pi$ se obtiene al despejar:
$$\pi = 4 * \left(\sum_{n=0}^{\infty}\frac{(-1)^{n}}{(2n+1)} \right)$$

La siguiente celda de codigo ofrece el valor de $\frac{\pi}{4}$

````python
import numpy as np
#El valor de pi/4
print("El valor de pi/4 es aproximadamente {}".format(np.pi/4))
````

````python
n=0
pi=0
while n<=100000:        #Indicamos de que valor a que valor haremos la suma
  pi+=4*(-1)**n/(2*n+1) #Vamos sumando cada elemento dentro de pi
  n+=1                  #Aumentamos el contador 1

print('El valor aproximado de pi={} y el de pi/4={}'.format(pi,pi/4))
````

Los métodos que pudiesen presentar errores como division entre cero, deben ser manejados de tal manera que el programa no imprima ningun error (vimos una clausula para hacer esto).

Adicional a esto, los métodos modulo y potencia deben ser generales, es decir, deben funcionar para cualquier número entero positivo (es un parámetro extra del método). Por ejemplo, la potencia de 2 elevado a 3 es 8, la potencia de 5 elevado a 3 es 125, etc.

Crea una instancia de la clase calculadora y prueba cada uno de los métodos, con valores de tu elección. Para los métodos que pudiesen presentar errores da dos ejemplos, uno totalmente funcional (3/5 por ejemplo) y otro que sea conflictivo pero que no regresa ningún error (4/0 por ejemplo).

````python
class Calculadora():
  def __init__(self, marca):
    self.marca = marca

  def suma(self, a, b):
    a = float(a) #Todos los valores los convertiremos a flotantes
    b = float(b)
    return a+b

  def resta(self, a, b):
    a = float(a)
    b = float(b)
    return a-b

  def producto(self, a, b):
    a = float(a)
    b = float(b)
    return a*b

  def division(self, a, b):
    a = float(a)
    b = float(b)
    if b != 0:
      return a/b
    else:
      print('Esta operación no se puede realizar:(')

  def modulo(self, a, b):
    a = int(a) #Usamos números enteros
    b = int(b)
    if b != 0:
      return a%b
    else:
      print('Esta operación no se puede realizar:(')

  def potencia(self, a, b):
    a = abs(int(a)) #Enteros positivos
    b = abs(int(b))
    return a**b
````

````python
casio = Calculadora('casio')
#Prueba de todas las operaciones
a= input('Ingrese un primer valor ')
b= input('Ingrese un segundo valor ')
print('Suma: {}'.format(casio.suma(a,b)))
print('Resta: {}'.format(casio.resta(a,b)))
print('Multiplicación: {}'.format(casio.producto(a,b)))
print('División: {}'.format(casio.division(a,b)))
print('Módulo: {}'.format(casio.modulo(a,b)))
print('Potencia: {}'.format(casio.potencia(a,b)))
````

Prueba la función con una lista de números de tu elección.

````python
def maximo(numeros):
  grande = numeros[0]
  for i in range(1, len(numeros)): #Para recorrer los valores de la lista
    if grande < numeros[i]: #Comparar los valores
      grande = numeros[i] #Nos quedamos con el mayor
  return grande
````

````python
def dado():
  return random.randint(1,6)
  
#Creamos una lista vacía en donde estarán los lanzamientos
lanzamientos_6 = []
i=0
while i<6: #Condición para que lo repita 6 veces
  lanzamientos_6.append(dado()) #Agregamos un elemento
  i+=1 #Sumamos un valor

print(lanzamientos_6)
````

El resultado de ``` lanzamientos_6``` será difícilmente [1,2,3,4,5,6], pues como ya sabemos la probabilidad de que nos de 1 en la primera entrada es de $\frac{1}{6}$, y en la segunda entrada $\frac{1}{6}$ y así con el resto de entradas, sin embargo para obtener justo en ese orden las probabilidades de cada evento se multiplican entonces tendríamos $(\frac{1}{6})^6=2.1433\times10^{-5}$ o $2.1433\times10^{-3}%$ de que este evento en específico suceda.
